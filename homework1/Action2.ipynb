```python
import numpy as np
import pandas as pd
```

```python
#读取成绩数据
data =pd.read_excel('grade.xlsx')
data
```

```python
#求统计值：平均成绩、最小/大成绩、方差、标准差等
data.describe()
```

```python
#计算各人总成绩
data['total'] = [data.iloc[0,1:-1].sum(),data.iloc[1,1:-1].sum(),data.iloc[2,1:-1].sum(),data.iloc[3,1:-1].sum(),data.iloc[4,1:-1].sum()]
data
```

```python
#按总成绩降序排名
df = data.sort_values('total',ascending=False)
df
```

```python
#重置索引
df.reset_index().drop(columns=['index'])
```
