```python
import numpy as np
import pandas as pd
```

```python
#读取数据
df = pd.read_csv('car_complain.csv')
df
```

```python
#拆分problem字段
for i in range(600):
    problems = str(df['problem'][i])
    num = problems.count(',')
    problems_list = problems.split(',')
    problems_list.pop()
    for j in problems_list:
        df.loc[i,j] = 1
df = df.fillna(0)
df
```

```python
#将’一汽-大众‘改为’一汽大众’，避免后续统计错误
df.brand[df['brand']=='一汽-大众'] = '一汽大众'
df
```

```python
#各品牌投诉总数
df1 = df.groupby(['brand']).count()
df1.id
```

```python
#各车型投诉总数
df2 = df.groupby(['car_model']).count()
df2.id
```

```python
#各品牌车型数
df3 = df.groupby(['brand','car_model']).count().groupby('brand').count()
df3.id
```

```python
#各品牌的平均车型投诉数及其排序
(df1.id/df3.id).sort_values(ascending=False)
```
